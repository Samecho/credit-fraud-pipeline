```python
import pandas as pd

creditcard = pd.read_csv("../data/raw/creditcard.csv")
```

```python
print(creditcard.shape)
creditcard.head()
```

```python
creditcard.info()
creditcard.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
creditcard["Class"].value_counts()
```

```python
sns.countplot(x='Class', data=creditcard)
plt.title('Histogram of Fraud')
plt.xlabel('Class')
plt.ylabel('Amount')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
creditcard["Scaled_Amount"] = scaler.fit_transform(creditcard[['Amount']])
creditcard["Scaled_Time"] = scaler.fit_transform(creditcard[['Time']])
creditcard[['Scaled_Amount', 'Scaled_Time']].head()
```

```python
y = creditcard["Class"]

x = creditcard.drop(["Class","Time","Amount"], axis=1)

x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, 
    y, 
    test_size=0.2, 
    random_state=0, 
    stratify=y
)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred[:5]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
weighted_model = RandomForestClassifier(
    n_estimators=100, 
    random_state=0, 
    n_jobs=-1, 
    class_weight='balanced'
)
```

```python
weighted_model.fit(x_train, y_train)
```

```python
y_pred_weighted = weighted_model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_weighted))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=0)
```

```python
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
print(y_train.value_counts())
print(y_train_smote.value_counts())
```

```python
smote_model = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
smote_model.fit(x_train_smote, y_train_smote)
```

```python
y_pred_smote = smote_model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_smote))
```
